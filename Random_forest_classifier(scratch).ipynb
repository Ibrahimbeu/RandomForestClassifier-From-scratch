```python
class Random_Forest():
    def __init__(self,n_estimators=5,max_depth=5,min_samples_split=2,max_features='sqrt'):
        self.np = __import__('numpy')
        self.num_trees=n_estimators
        self.max_depth= max_depth
        self.max_features= max_features
        self.min_samples_split= min_samples_split
        self.base_model= __import__("Decision_Tree").Decision_Tree
        self.trees= []
    def samples(self,dft):
        temp =dft.iloc[:,[i for i in range(len(dft.columns)) if i!=len(dft.columns)-1]].copy()
        index= self.np.random.choice(temp.index,size=temp.shape[0],replace=True)
        
        if self.max_features=='sqrt':
            number=int(len(temp.columns)**0.5)
        elif self.max_features=='log2':
            number= self.np.log2(len(temp.columns))
        elif self.max_features==None:
            number=len(temp.columns)
        
        cols= self.np.random.choice(temp.columns,size=number)
        return dft.loc[index,cols]
    def fit(self,dft):
        if len(self.trees)>0:
            self.trees=[]
        for _ in range(self.num_trees):
            dft_= Random_Forest().samples(dft)
            model= self.base_model(min_samples=self.min_samples_split,max_depth= self.max_depth)
            print(model)
#             x_=dft.iloc[:,[i for i in range(len(dft.columns)) if i!=len(dft.columns)-1]]
#             y_= dft.iloc[:,-1]
            model.fit(dft_)
            self.trees.append(model)
    def predict(self,x):
        predictions= []
        predict_final=[]
        for mdl in self.trees:
            predictions.append(mdl.predict(x))
        predictions= self.np.transpose(self.np.array(predictions))
        for pr in predictions:
            val,cnt= self.np.unique(pr,return_counts=True)
            predict_final.append(val[self.np.argmax(cnt)])
        return predict_final
```
